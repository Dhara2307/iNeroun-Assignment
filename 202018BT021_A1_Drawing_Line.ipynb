```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

```

```python
img = np.ones((512,512))
img = img.astype('uint8')
```

```python
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
plt.xlabel('x')
plt.ylabel('y')
```

```python
line_img = img
x1=0; y1=512; x2=512; y2=0;
cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)
```
